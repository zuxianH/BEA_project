# Define Log function

```
log[z_, branch_ : 0, branchcut_ : -Pi] := 
  Log[Abs[z]] + 
   I*(Arg[z Exp[-I (branchcut + Pi)]] + branchcut + Pi + 
      2 Pi*branch);
F[u_, branch_ : 0, 
   branchcut_ : -Pi] := (-1/(2 Pi*I)) log[(u + I)/(u - I), branch, 
    branchcut];
Fsym[u_, branch_ : 0, branchcut_ : -Pi] := 
  F[u, branch, branchcut] - 1/2;
```

## Plot F

```
Plot[{F[u], Fsym[u], Fsym[u, -1, 0]}, {u, -Infinity, Infinity}, 
 PlotRange -> {{-Infinity, Infinity}, {-0.7, 1}}]
```

# Iterative solver with loops construction While 

```
uZeroEpsilon[k_, M_, L_, nk_] := (1/2) Tan[-(\[Pi] (nk))/L];
BAELog1[k_, L_, M_, \[Epsilon]_, nk_] := 
 L Fsym[2 u[k], -1, 
    0] == (\[Epsilon]) Sum[
     Fsym[(u[k] - u[j])/1, -1, 0] Boole[j != k], {j, 1, M}] + nk
BAELog2[k_, L_, M_, \[Epsilon]_, nk_] := 
 L Fsym[2 u[k], -1, 
    0] == (1) Sum[
     Fsym[(u[k] - u[j])/\[Epsilon], -1, 0] Boole[j != k], {j, 1, M}] +
    nk
BAELog3[k_, L_, M_, \[Epsilon]_, nk_] := 
 L Fsym[2 u[k], -1, 
    0] == (\[Epsilon]) Sum[
     Fsym[(u[k] - u[j])/1, -1, 0] Boole[j != k], {j, 1, M}] + nk


BAElogsolver[L_, M_, \[Epsilon]_, num_, uInitial_] := 
 FindRoot[
  Table[BAELog1[k, L, M, \[Epsilon], nk[[num]][[k]]], {k, 1, M}], 
  Table[{u[k], uInitial[[k]]}, {k, 1, M}]]

(*Main iterative solver*)
IterativeBAEsolver[L_, M_, num_, \[Epsilon]Step_, \[Epsilon]Max_] := 
  Module[{epsilon = 1, solution, uInitial, 
    results},(*Initialize the first guess_ for uInitial*)
   uInitial = 
    Table[uZeroEpsilon[k, M, L, nk[[num]][[k]]], {k, 1, M}];
   results = {};
   While[epsilon <= \[Epsilon]Max,(*Solve for the current epsilon*)
    solution = BAElogsolver[L, M, epsilon, num, uInitial] // Quiet;
    AppendTo[results, {epsilon, solution // Values}];
    (*Update the initial guess for the next iteration*)
    uInitial = solution // Values;
    (*Increment epsilon*)epsilon += \[Epsilon]Step;];
   results];


L = 10;
M = 4;
nk = Subsets[Range[-((L - (M - 1))/2) + 1, (L - (M - 1))/2 - 1], {M}];

result = 
  Table[{nk[[nkvalue]], 
      IterativeBAEsolver[L, M, nkvalue, 0.01, 1][[-1]][[2]]} // 
     Re, {nkvalue, Length[nk]}] // Timing;
result
```

# Iterative solver with loops construction ParallelTable 

```



uZeroEpsilon[k_, M_, L_, nk_] := (1/2) Tan[-(\[Pi] (nk))/L];
BAELog[k_, L_, M_, \[Epsilon]_, nk_] := 
    L Fsym[2 u[k], -1, 0] == (\[Epsilon]) Sum[
           Fsym[u[k] - u[j], -1, 0] Boole[j != k], {j, 1, M}] + nk;


BAElogsolver[L_, M_, \[Epsilon]_, num_, uInitial_] := 
    Quiet@FindRoot[
        Table[BAELog[k, L, M, \[Epsilon], nk[[num]][[k]]], {k, 1, M}], 
        Table[{u[k], uInitial[[k]]}, {k, 1, M}]
      ];

(* Main iterative solver with parallel computation *)
IterativeBAEsolver[L_, M_, num_, \[Epsilon]Step_, \[Epsilon]Max_] := 
    Module[
       {epsilonList, initialGuesses, results, parallelResults},
       
       (* Generate a list of epsilon values *)
       epsilonList = Range[1/100, \[Epsilon]Max, \[Epsilon]Step];
       
       (* Initialize the first guess for uInitial *)
       initialGuesses = 
          Table[uZeroEpsilon[k, M, L, nk[[num]][[k]]], {k, 1, M}];
       
       (* Parallel computation for solving BAE at each epsilon *)
       parallelResults = ParallelTable[
           Module[{epsilon, solution},
              epsilon = eps;
              
      solution = BAElogsolver[L, M, epsilon, num, initialGuesses];
              initialGuesses = solution // Values; (* 
      Update for next iteration *)
              {epsilon, initialGuesses}
            ],
           {eps, epsilonList}
         ];
       
       (* Extract results *)
       results = parallelResults;
       results
     ];

L = 25;
M = 10;
nk = Subsets[Range[-((L - (M - 1))/2) + 1, (L - (M - 1))/2 - 1], {M}];
result = 
  Table[{nk[[nkvalue]], 
      IterativeBAEsolver[L, M, nkvalue, 0.01, 1][[-1]][[2]]} // 
     Re, {nkvalue, Length[nk]}] // Timing;
```
